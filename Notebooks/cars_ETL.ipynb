Import libraries

```python
import pandas as pd; pd.set_option('display.max_columns', 100)
import numpy as np
```

Load Data

```python
#read data
df = pd.read_csv('../Data/ML_cars.csv')
```

Visualize the data

```python
#Preview tipe of data 
def summary(df):
    print(f'data shape: {df.shape}')
    summ = pd.DataFrame(df.dtypes, columns=['data type'])
    summ['#missing'] = df.isnull().sum()
    summ['%missing'] = df.isnull().sum()*100/len(df)
    summ['#unique'] = df.nunique()
    desc = pd.DataFrame(df.describe(include='all').T)
    summ['min'] = desc['min']
    summ['max'] = desc['max']
    return summ

summary(df)
```

Check the unique values ​​of some columns

```python
unique_risk= df['symboling'].unique()
unique_fuel= df['fueltype'].unique()
unique_aspiration= df['aspiration'].unique()
unique_door= df['doornumber'].unique()
unique_carbody= df['carbody'].unique()
unique_drivewheel= df['drivewheel'].unique()
unique_loc= df['enginelocation'].unique()
unique_type= df['enginetype'].unique()
unique_cylinder= df['cylindernumber'].unique()
unique_fuelsystem= df['fuelsystem'].unique()	

print('Risk Value:', unique_risk)
print('Fuel type:', unique_fuel)
print('Aspiration:', unique_aspiration)
print('Doors number:', unique_door)
print('Car Body', unique_carbody)
print('Drive wheels:', unique_drivewheel)
print('Engine location:', unique_loc)
print('Engine type:', unique_type)
print('Numner of cylinders:', unique_cylinder)
print('Fuel System:', unique_fuelsystem)
```

```python
# Split the "CarName" column into two parts (Brand and model) and select the brand
df[['Brand', 'Model']] = df['CarName'].str.split(' ', n=1, expand=True)

# dropping the 'CarName' column from the DataFrame
df = df.drop('CarName', axis=1)
# Get the unique values ​​of the "Brand" column
unique_brands = df['Brand'].unique()

# Print the unique values ​​of the brand
print("Unique Brands:")
print(unique_brands)
```

```python
# Capitalize the first letter of each word in the new columns
df['Brand'] = df['Brand'].str.title()
df['Model'] = df['Model'].str.title()

# Print the DataFrame with the new columns
print(df[['Brand', 'Model']])
```

Reorder data

```python
# Get the "car_ID" index
car_id_index = df.columns.get_loc("car_ID")

# Reorder columns
new_columns_order = list(df.columns[:car_id_index]) + ['Brand', 'Model'] + list(df.columns[car_id_index:-2])

# Create new dataframe with changes
df_reordered = df[new_columns_order]

df = df_reordered
df = df.drop('car_ID', axis=1)
```

Correct wrong values in Brand Column

```python
# Replace incorrect names in "Brand" column
corrections = {
    'Alfa-Romero': 'Alfa Romeo',
    'Bmw': 'BMW',
    'Maxda': 'Mazda',
    'nissan': 'Nissan',
    'Porcshce': 'Porsche',
    'Toyouta': 'Toyota',
    'Vokswagen': 'Volkswagen',
    'Vw': 'Volkswagen'
}

df['Brand'] = df['Brand'].replace(corrections)

# Get the unique values ​​of the "Brand" column
unique_brands = df['Brand'].unique()

# Print the unique values ​​of the brand
print("Unique Brands:")
print(unique_brands)
```

Models per Brand

```python
# Get the relationship between make and model
brand_model_relation = df.groupby('Brand')['Model'].unique().reset_index()

# Print
print(brand_model_relation)
```
